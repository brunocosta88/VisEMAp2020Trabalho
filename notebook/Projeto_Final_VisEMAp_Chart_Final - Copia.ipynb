```python
# Import packages
import numpy as np
import pandas as pd
import altair as alt
from altair_saver import save
```

```python
# Load dataset
df3 = pd.read_csv('C:/Users/MediaTechLab/Documents/GitHub/VisEMAp2020Trabalho/data/video_sessions_20201210-20201216_2.csv')
#df3.qualidade = pd.to_numeric(df3.qualidade)
df3 = df3.drop(['Unnamed: 0'], axis = 1)
df3.head(3)
```

```python
# Heatmap da variação percentual da média móvel 

# Título geral
title = alt.Chart(
    {"values": [{"text": "The Title"}]}
).mark_text(size=20).encode(
    text="text:N"
)

# radio/dropdown/slider https://altair-viz.github.io/gallery/multiple_interactions.html e alterar posição https://stackoverflow.com/questions/62103632/altair-change-the-position-of-a-slider 
regiao_radio = alt.binding_radio(options=np.append([None],df3['device'].unique()),labels=['Todas'])
selection = alt.selection_single(fields=['device'], bind=regiao_radio, name="Selecione")

# gráfico das taxas de desemprego
# Tooltip https://vegawidget.github.io/altair/articles/tooltips.html
points = alt.Chart(data= df3, width=800, height=300, title='Monitoração Bitrate VOD Globoplay, 10/12 a 16/12').mark_rect(filled=True).encode(
    x= alt.X('qualidade', type='ordinal', title= 'Taxa de Video'),
    y = alt.Y('device', title=None, sort = alt.SortField("device", order = "descending")),
    tooltip = [alt.Tooltip(field = "player", title = "Player", type = "nominal"), 
                alt.Tooltip(field = "device", title = "Device", type = "nominal"),
                alt.Tooltip(field = "qualidade",title = "Bitrate", type = "ordinal"),
                alt.Tooltip(field = "percentual",title = "Percentual de Impressões", type = "quantitative"),
                alt.Tooltip(field = "impressoes",title = "Número de Impressões", type = "quantitative")],
    color=alt.condition(selection, 
                        alt.Color('impressoes:Q', 
                                  scale=alt.Scale(scheme='blues', 
                                                  domain=[0, max(df3.impressoes)]),
                                  legend=alt.Legend(direction='vertical',
                                                    orient='left', legendY=30,
                                                    title = 'Impressões',
                                                    titleOrient = 'left', titleFontSize = 10,
                                                    titleFontWeight = 'normal')),
                        alt.value('lightgray'))
  ).add_selection(
    selection
  )

text=alt.Chart().mark_text(align='center', baseline='middle',color='black').encode(
    y=alt.Y('device:N', title=None),
    x='qualidade:O',
    #color=alt.Color('impressoes:Q', legend=None, scale=alt.Scale(range=['black'])),
    text='impressoes:Q')

chart = alt.hconcat(alt.layer(points,text, data = df3), spacing=0).resolve_legend()

chartfinal = chart

chartfinal
#save(chartfinal, "chartfinal2.html")

# Depois do gráfico salvo, adicionar seguinte código no HTML conforme https://stackoverflow.com/questions/62103632/altair-change-the-position-of-a-slider
#<style>
#form.vega-bindings {
#  position: absolute;
#  right: 568px;
#  top: 63px;
#}
#</style>
```

```python
# Load dataset
df = pd.read_csv('C:/Users/MediaTechLab/Documents/GitHub/VisEMAp2020Trabalho/data/video_sessions_20201210-20201216.csv')
#df.qualidade = pd.to_numeric(df.qualidade)
df2 = df.drop(['Unnamed: 0'], axis = 1)
df2.head(3)
```

```python
df4 = df2[df2.device == 'tv']
df4.head(3)
```

```python
#Disabling Large Row Error

alt.data_transformers.disable_max_rows()

# Heatmap da variação percentual da média móvel 

# Título geral
title = alt.Chart(
    {"values": [{"text": "The Title"}]}
).mark_text(size=20).encode(
    text="text:N"
)

h0 = len(np.unique(df4.player))

# gráfico das taxas de desemprego
# Tooltip https://vegawidget.github.io/altair/articles/tooltips.html
points = alt.Chart(data= df4, width=800, height=30*h0, title='Bitrate VOD Globoplay (TVs), 10/12 a 16/12').mark_rect(filled=True).encode(
    x= alt.X('qualidade', type='ordinal', title= 'Taxa de Video'),
    y = alt.Y('player', title=None, sort = alt.SortField('player', order = "descending")),
    tooltip = [alt.Tooltip(field = "player", title = "Player", type = "nominal"), 
                alt.Tooltip(field = "device", title = "Device", type = "nominal"),
                alt.Tooltip(field = "qualidade",title = "Bitrate", type = "ordinal"),
                alt.Tooltip(field = "percentual",title = "Percentual de Impressões", type = "quantitative"),
                alt.Tooltip(field = "impressoes",title = "Número de Impressões", type = "quantitative")],
    color= alt.Color('impressoes:Q', scale=alt.Scale(scheme='blues', domain=[0, max(df4.impressoes)]), 
                          legend=alt.Legend(direction='vertical', orient='left', legendY=30, title = 'Impressões', titleOrient = 'left', titleFontSize = 10, titleFontWeight = 'normal'))
  )

text=alt.Chart().mark_text(align='center', baseline='middle',color='black').encode(
    y=alt.Y('player:N', title=None, sort = alt.SortField("device", order = "ascending")),
    x='qualidade:O',
    #color=alt.Color('impressoes', legend=None, scale=alt.Scale(range=['white'])),
  # labelFontSize=15,
    text='impressoes:Q')

chart = alt.hconcat(alt.layer(points,text, data = df4), spacing=0).resolve_legend()


chartfinal = chart

chartfinal
#save(chartfinal, "chartfinal2.html")

# Depois do gráfico salvo, adicionar seguinte código no HTML conforme https://stackoverflow.com/questions/62103632/altair-change-the-position-of-a-slider
#<style>
#form.vega-bindings {
#  position: absolute;
#  right: 568px;
#  top: 63px;
#}
#</style>
```
